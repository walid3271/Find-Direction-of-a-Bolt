```python
img = cv2.imread(image_path)
points = predict[0].obb.xyxyxyxy[0].numpy().astype(int)
cv2.polylines(img, [points], isClosed=True, color=(255, 255, 0), thickness=1)

xmidtop = int((points[0][0] + points[1][0]) / 2)
ymidtop = int((points[0][1] + points[1][1]) / 2)

xmidbot = int((points[2][0] + points[3][0]) / 2)
ymidbot = int((points[2][1] + points[3][1]) / 2)

xmidlft = int((points[0][0] + points[3][0]) / 2)
ymidlft = int((points[0][1] + points[3][1]) / 2)

xmidrit = int((points[1][0] + points[2][0]) / 2)
ymidrit = int((points[1][1] + points[2][1]) / 2)


xmidmid = int((xmidtop + xmidbot) / 2)
ymidmid = int((ymidtop + ymidbot) / 2)



cv2.line(img, (xmidtop, ymidtop), (xmidbot, ymidbot), (0, 0, 255), 2)
cv2.line(img, (xmidlft, ymidlft), (xmidrit, ymidrit), (0, 255, 0), 2)

cv2.line(img, (points[0][0]-100, ymidmid), (points[2][0]+100, ymidmid), (20, 20, 20), 2)
cv2.line(img, (xmidmid, points[0][1]), (xmidmid, points[2][1]), (20, 20, 20), 2)

cv2.line(img, (xmidbot, ymidbot), (xmidmid, ymidmid), (255, 0, 255), 2)


cv2_imshow(img)
```

```python
angle = -1*(np.arctan2(ymidbot - ymidmid, xmidbot - xmidmid) * 180 / np.pi)
print(angle)
```
