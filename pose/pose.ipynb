```python
from ultralytics import YOLO
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image_path = "/content/drive/MyDrive/HawkEyes/japan/bolt/Image__2024-03-27__13-43-34.png"
```

```python
model = YOLO("/content/drive/MyDrive/HawkEyes/japan/pose/best.pt")
```

```python
predict = model.predict(image_path , save = True , save_txt = True)
```

```python
points = predict[0].keypoints.xy[0].numpy().astype(int)
```

```python
xmid = int((points[0][0]+points[1][0])/2)
ymid = int((points[0][1]+points[1][1])/2)
```

```python
img = cv2.imread(image_path)
cv2.line(img, (points[0][0], points[0][1]), (xmid, ymid), (255, 0, 0), 2)
cv2.line(img, (xmid, ymid), (points[1][0], points[1][1]), (0, 0, 255), 2)
cv2_imshow(img)
```
